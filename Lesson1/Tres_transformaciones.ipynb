```python
A = np.array([[3, 7], [5, 2]])
A
```

```python
print('Circulo unitario')

graficarMatriz(np.array([[1, 0], [0, 1]]))

plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)

plt.show()
```

```python
print('Circulo unitario transformado')

graficarMatriz(A)

plt.xlim(-8, 8)
plt.ylim(-8, 8)

plt.show()
```

```python
U, D, V = np.linalg.svd(A)
print(U)
```

```python
print('Circulo unitario')

graficarMatriz(np.array([[1, 0], [0, 1]]))

plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)

plt.show()

print('Rotacion (V):')

graficarMatriz(V)

plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)

plt.show()

print('Escala (D):')

graficarMatriz(np.diag(D).dot(V))

plt.xlim(-9, 9)
plt.ylim(-9, 9)

plt.show()

print('Segunda rotacion (U):')

graficarMatriz(U.dot(np.diag(D).dot(V)))

plt.xlim(-8, 8)
plt.ylim(-8, 8)

plt.show()
```

```python
# Aplicar las transformaciones U, D, V por separado es lo mismo que aplicar A
print('Circulo unitario transformado')

graficarMatriz(A)

plt.xlim(-8, 8)
plt.ylim(-8, 8)

plt.show()
```
